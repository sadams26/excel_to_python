# Translating Basic Excel to Python


## Importing

```python
import pandas as pd
```

## Read Excel File and Assign to Variable Named `df`

```python
df = pd.read_excel("https://raw.githubusercontent.com/sadams26/excel_to_python/main/data/AB_NYC_2019.xlsx")
```

## View First Five Rows with `head`

```python
df.head()
```

## Display First Five Rows with Indexing

```python
df[0:5]
```

## Display First Five Rows and Select Columns with `loc`

```python
df.loc[0:5, ["id", "name", "host_id", "host_name"]]
```

## Display All Rows and Select Columns

```python
df.loc[:, ["id", "name", "host_id", "host_name"]]
```

## Display All Rows and All Columns

```python
df.loc[:,]
```

## Display All Rows Where the Neighbourhood Group is Brooklyn

```python
df.loc[:, "price"]
```

```python
df.loc[df["neighbourhood_group"]=="Brooklyn", ]
```

## Display All Rows Where the Neighbourhood Group is Brooklyn or Manhattan

```python
df.loc[df["neighbourhood_group"].isin(["Brooklyn", "Manhattan"]), ]
```

## Display All Rows Where the Neighbourhood Group is **NOT** Brooklyn

```python
df.loc[df["neighbourhood_group"]!="Brooklyn", ]
```

## Output the Mean Price

```python
df.loc[:, ["price"]].mean()
```

## Output Multiple Descriptive Measures with `describe`

```python
df.loc[:, ["price"]].describe()
```

## Create a Pivot Table

```python
pd.pivot_table(df, values="price", index="neighbourhood_group", aggfunc="mean", margins=True)
```
